```python
import pandas as pd
data = {
    'people': ['Karan', 'Abhishek', 'Vinesh', 'Vardhan'],
    'age': [20, 24, 18, 40],
    'height': [175, 160, 165, 155],
    'weight': [75, 85, 60, 89]
}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(df['age'], df['height'], s=100, color='blue', label='Height')
plt.scatter(df['age'], df['weight'], s=100, color='red', label='Weight')
plt.title('Relationship between Age and Health Metrics')
plt.xlabel('Age')
plt.ylabel('Height / Weight')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df['people'], df['height'], color='blue', label='Height')
plt.bar(df['people'], df['weight'], color='red', label='Weight')
plt.title('Comparison of Height and Weight among Individuals')
plt.xlabel('People')
plt.ylabel('Height / Weight')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['height'], bins=5, color='blue', alpha=0.8, label='Height')
plt.hist(df['weight'], bins=5, color='red', alpha=0.6, label='Weight')
plt.title('Distribution of Height and Weight')
plt.xlabel('Measurement')
plt.ylabel('Frequency')
```
